```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv('/content/spam.csv', encoding="ISO-8859-1")
df.sample(7)
```

```python
df = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1)
```

```python
df = df.rename(columns={'v1':'target','v2':'Mails'})
```

```python
df.sample(7)
```

```python
df['target'].unique()
```

```python
df['target'] = df['target'].map({'ham':1 , 'spam':0}).astype(int)
```

```python
import string
string.punctuation
```

```python
import nltk
nltk.download('stopwords')
```

```python
import nltk
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
le = WordNetLemmatizer()
le.lemmatize('care')
```

```python
text_cleaned_data = []

def clean_data(i,mode):
  if mode == 'train':
    text = i.lower() # lowering the data
    text = ''.join([j for j in text if j not in string.punctuation]) # remove punctuations
    text = ' '.join([k for k in text.split() if k not in stopwords.words('english')]) # remove stopwords
    text = ' '.join([le.lemmatize(p) for p in text.split()])
    text_cleaned_data.append(text)
  elif mode == 'inference':
    text = i.lower() # lowering the data
    text = ''.join([j for j in text if j not in string.punctuation]) # remove punctuations
    text = ' '.join([k for k in text.split() if k not in stopwords.words('english')]) # remove stopwords
    text = ' '.join([le.lemmatize(p) for p in text.split()])
    return text
  else:
    print('Please Pass Mode')

for i in df['Mails']:
  clean_data(i,'train')
```

**Bag_of_words -> ML**

`Giving Independent data which is already cleaned to bag of words (so that it will convert into numbers) -> then give to ML `

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
cv.fit(text_cleaned_data)
bow_res = cv.transform(text_cleaned_data).toarray()
bow_res
```

```python
from sklearn.naive_bayes import GaussianNB
reg_bow = GaussianNB()
reg_bow.fit(bow_res,df['target'])
```

**Accuracy_BOW**

```python
from sklearn.metrics import accuracy_score
print(f'Accuracy with BOW -> : {accuracy_score(df["target"],reg_bow.predict(bow_res))}')
```

```python
import pickle
with open('BOW_ML.pkl','wb') as f:
  pickle.dump(reg_bow,f)
```

```python
# Load file and do predictions

with open('/content/BOW_ML.pkl','rb') as f:
  model = pickle.load(f)
```

```python
with open('BOW_CV.pkl','wb') as f:
  pickle.dump(cv,f)
```

```python
with open('/content/BOW_CV.pkl','rb') as f:
  tonum = pickle.load(f)
```

```python
# Testing:
import warnings
warnings.filterwarnings('ignore')

john_mail = 'I watched a Movie and it was really fantastic'

outcome_ = clean_data(john_mail,'inference')

input_ = tonum.transform([outcome_]).toarray()

r = model.predict(input_)[0]
if r == 1:
  print(f'John Main will come to Inbox ')
else:
  print(f'John Main will be in Spam Folder')
```

**EMbedding -> ML**

```python
c = []
for i in text_cleaned_data:
  c.append(len(i.split()))
print(max(c))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
dic_size = 1000

result = [one_hot(j,dic_size) for j in text_cleaned_data]
data_input = pad_sequences(result,80,padding='post')
data_input
```

```python
from sklearn.naive_bayes import GaussianNB
em_bow = GaussianNB()
em_bow.fit(data_input,df['target'])
```

**Accuracy_BOW**

```python
from sklearn.metrics import accuracy_score
print(f'Accuracy with BOW -> : {accuracy_score(df["target"],em_bow.predict(data_input))}')
```
